```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initialise figure instance
fig, ax = plt.subplots(1, 1, dpi=150)

# Plot
x = np.linspace(0, 2*np.pi, 100)
ax.plot(x, np.sin(x))
ax.plot(x, np.sin(x+0.5*np.pi))
ax.plot(x, np.sin(x+1.0*np.pi))
ax.plot(x, np.sin(x+1.5*np.pi))

ax.set_xlim(0, 2 * np.pi)
ax.set_title("Sineusoids") #, color=COLOR_SECONDARY)
ax.set_xlabel(r'$\theta$')
ax.set_ylabel(r'$\sin (\theta)$');
```

```python
## LaTeX and Styling example
import mpl_helper
plt.style.use("example.mplstyle")


# Initialise figure instance
fig, ax = plt.subplots(1, 1, figsize=mpl_helper.set_size(252.0), dpi=150)

# Plot
x = np.linspace(0, 2*np.pi, 100)
ax.plot(x, np.sin(x))
ax.plot(x, np.sin(x+0.5*np.pi))
ax.plot(x, np.sin(x+1.0*np.pi))
ax.plot(x, np.sin(x+1.5*np.pi))

ax.set_xlim(0, 2 * np.pi)
ax.set_title("Sineusoids")#, color=f'#{mpl_helper.COLOR_PRIMARY}')
ax.set_xlabel(r'$\theta$')
ax.set_ylabel(r'$\sin (\theta)$');

fig.savefig('example_1.pdf', format='pdf', bbox_inches='tight')
```
